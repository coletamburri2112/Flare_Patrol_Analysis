```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
from astropy.io import fits
```

```python
filename_qs_CaII = '/Users/coletamburri/Desktop/nonflareCaII.npz'
filename_qs_Hbeta = '/Users/coletamburri/Desktop/nonflareHbeta.npz'
```

```python
CaII_qs = np.load(filename_qs_CaII)
Hbeta_qs = np.load(filename_qs_Hbeta)
```

```python
CaII_qs
```

```python
nf_caII = CaII_qs['nonflare']
wl_caII = CaII_qs['wl']

nf_hbeta = Hbeta_qs['nonflare']
wl_hbeta = Hbeta_qs['wl']
```

```python
avg_CaIIqs = np.mean(nf_caII,1)
```

```python
avg_Hbetaqs = np.mean(nf_hbeta,1)
```

```python
dkist_coord_file = '/Users/coletamburri/Desktop/11_Aug_2024_Cclass_Flare/Processed_ViSP_VBI_11Aug2024/ViSPcoords.npz'
dkist_coords = np.load(dkist_coord_file)

xarr_caII = dkist_coords['xarr_caII']
yarr_caII = dkist_coords['yarr_caII']

xarr_hbeta = dkist_coords['xarr_hbeta']
yarr_hbeta = dkist_coords['yarr_hbeta']
```

```python
yarr_caII[1900]
```

```python
startind = 1900
yarr_caII_nf = yarr_caII[startind:]
yarr_hbeta_nf = yarr_hbeta[startind:]
```

```python
space_and_time_averaged_nf_CaII = np.mean(np.mean(nf_caII,2),0)
space_and_time_averaged_nf_Hbeta = np.mean(np.mean(nf_hbeta,2),0)
```

```python
fig,[(ax,ax2),(ax3,ax4)]=plt.subplots(2,2,dpi=200,figsize=(10,10))
ax.pcolormesh(xarr_caII,yarr_caII_nf,np.transpose(avg_CaIIqs),vmin=0,vmax=3e6,cmap='grey')
ax2.pcolormesh(xarr_hbeta[:-1],yarr_hbeta_nf,np.transpose(avg_Hbetaqs),vmin=0,vmax=6e6,cmap='grey')
ax.set_aspect('equal')
ax2.set_aspect('equal')
ax3.plot(wl_caII,space_and_time_averaged_nf_CaII,color='#CC6677')
ax4.plot(wl_hbeta,space_and_time_averaged_nf_Hbeta,color='#AAAAAA')
ax3.set_xlabel('Wavelength [nm]')
ax4.set_xlabel('Wavelength [nm]')
ax3.set_ylabel(r'Intensity [$\text{erg s}^{-1}\;\text{cm}^{-2}\;\AA^{-1}\;\text{sr}^{-1}$]')
ax4.set_ylabel(r'Intensity [$\text{erg s}^{-1}\;\text{cm}^{-2}\;\AA^{-1}\;\text{sr}^{-1}$]')
ax3.axvline(396.847,c='black',linestyle='dashed')
ax4.axvline(486.1375,c='black',linestyle='dashed')
ax3.axvline(397.01,c='grey',linestyle='dashed')

ax.set_xlabel('HPC-x [arcsec]')
ax2.set_xlabel('HPC-x [arcsec]')
ax.set_ylabel('HPC-y [arcsec]')
ax2.set_ylabel('HPC-y [arcsec]')



```
