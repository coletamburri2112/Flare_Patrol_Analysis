```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

```

```python
csvfile = '/Users/coletamburri/Desktop/11_Aug_2024_Cclass_Flare/OSPEX_Fit_Results_11Aug2024/fitting_comp.csv'
column_names = ['Fnew', 'dnew', 'cutnew','s', 'Fnew_Err','dnew_err','cutnew_err','s2', 'Fold','dold','cutold','s3', 'Fold_Err','dold_err','cutold_err']

names=column_names
```

```python
csv = pd.read_csv(csvfile,header=None,names=column_names)
```

```python
csv = csv.iloc[:11]
```

```python
csv
```

```python
fig,[ax,ax1,ax2]=plt.subplots(1,3,dpi=200,figsize=(15,3))
ax.errorbar(range(11), csv['Fnew'], yerr=csv['Fnew_Err'],label='[4-32 keV], [40-50 keV]',color='#CC6677',marker='.',capsize=2)
ax.errorbar(range(11), csv['Fold'], yerr=csv['Fold_Err'],label='[4-50 keV]',color='black',marker='.',capsize=2)
ax.set_ylabel(r'$F_e$ [$s^{-1}$]')
ax.set_xlabel('Fit interval')

ax.legend(loc='lower right')

ax1.errorbar(range(11), csv['dnew'], yerr=csv['dnew_err'],label='[4-32 keV], [40-50 keV]',color='#CC6677',marker='.',capsize=2)
ax1.errorbar(range(11), csv['dold'], yerr=csv['dold_err'],label='[4-50 keV]',color='black',marker='.',capsize=2)
ax1.set_ylabel(r'$\delta_{\;<\;E_b}$')
ax1.set_xlabel('Fit interval')

ax1.legend(loc='lower right')

ax2.errorbar(range(11), csv['cutnew'], yerr=csv['cutnew_err'],label='[4-32 keV], [40-50 keV]',color='#CC6677',marker='.',capsize=2)
ax2.errorbar(range(11), csv['cutold'], yerr=csv['cutold_err'],label='[4-50 keV]',color='black',marker='.',capsize=2)
ax2.set_ylabel(r'$E_{c,l}\; [keV]$')
ax2.set_ylim([5,25])
ax2.set_yticks([7,12,17,22,27])
ax2.set_xlabel('Fit interval')
ax2.legend(loc='lower right')
```

```python
fig,ax=plt.subplots()
ax.errorbar(range(11), csv['dnew'], yerr=csv['dnew_err'],label='new')
ax.errorbar(range(11), csv['dold'], yerr=csv['dold_err'],label='old')
ax.set_title(r'$\delta$')

ax.legend()
```

```python
fig,ax=plt.subplots()
ax.errorbar(range(11), csv['cutnew'], yerr=csv['cutnew_err'],label='new')
ax.errorbar(range(11), csv['cutold'], yerr=csv['cutold_err'],label='old')
ax.set_title(r'$E_c$')

ax.legend()
```
