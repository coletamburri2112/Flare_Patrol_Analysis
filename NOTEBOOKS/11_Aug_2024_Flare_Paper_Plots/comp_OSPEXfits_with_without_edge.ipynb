```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

```

```python
csvfile = '/Users/coletamburri/Desktop/fitting_comp.csv'
column_names = ['Fnew', 'dnew', 'cutnew','s', 'Fnew_Err','dnew_err','cutnew_err','s2', 'Fold','dold','cutold','s3', 'Fold_Err','dold_err','cutold_err']

names=column_names
```

```python
csv = pd.read_csv(csvfile,header=None,names=column_names)
```

```python
csv = csv.iloc[:11]
```

```python
csv
```

```python
fig,ax=plt.subplots()
ax.errorbar(range(11), csv['Fnew'], yerr=csv['Fnew_Err'],label='new')
ax.errorbar(range(11), csv['Fold'], yerr=csv['Fold_Err'],label='old')
ax.set_title('Elec. flux')
ax.legend()
```

```python
fig,ax=plt.subplots()
ax.errorbar(range(11), csv['dnew'], yerr=csv['dnew_err'],label='new')
ax.errorbar(range(11), csv['dold'], yerr=csv['dold_err'],label='old')
ax.legend()
```

```python
fig,ax=plt.subplots()
ax.errorbar(range(11), csv['cutnew'], yerr=csv['cutnew_err'],label='new')
ax.errorbar(range(11), csv['cutold'], yerr=csv['cutold_err'],label='old')
ax.legend()
```
