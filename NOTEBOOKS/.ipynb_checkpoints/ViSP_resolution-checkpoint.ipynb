```python
# first compute RMS intensity contract
def RMS_contrast(spec):
    rms_arr = []
    for i in range(np.shape(spec)[0]):
        arrch = spec[i,:]
        std = np.nanstd(arrch)
        mean = np.nanmean(arrch)
        rms_arr.append(100 * std/mean)
    return rms_arr
```

```python
# now do the FFT, one for each slit position (try with the 90th, first)
# perform hann window first
# then 1D FFT

def power_spec(spec):
    psd_arr = []
    for i in range(np.shape(spec)[0]):
        arrch = spec[i,:]
        hannlen = hann(len(arrch))
        windowed = arrch*hannlen
        fastfourt = fft(windowed)
        sq = fastfourt**2
        fs = 1/0.0295 #sampling frequency
        N = arrch.size
        psd = sq/fs/N
        psd_arr.append(windowed)
    step = 0.0295 #arm1
    freqs = np.fft.fftfreq(arrch.size, step)
    idx = np.argsort(freqs)
    return psd_arr, freqs, idx
```
