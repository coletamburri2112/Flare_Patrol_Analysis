```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
import dkistpkg_ct
```

```python
from scipy.optimize import curve_fit

def fitline(xdata, ydata, plot='True'):
	#Gaussian fit
	params, covariance = curve_fit(dkistpkg_ct.gaussian_for_calib, xdata, ydata, p0=[-0.2, np.mean(xdata), 1,0.4])
	a,b,c,d = params
	xfit  = np.linspace(xdata[0],xdata[-1], 1000)
	yfit = dkistpkg_ct.gaussian_for_calib(xfit, a,b,c,d)

	w0 = xfit[np.argmin(yfit)]
	print(' MINIMA loc--->',w0)
	
	if plot=='True':
		plt.plot(xdata, ydata, '.')
		plt.plot(xfit, yfit)
		plt.plot([w0,w0], [yfit.min()-0.1, yfit.max()], color='gray')
		plt.ylim(ydata.min()-0.03, ydata.max()+0.03)

	return xfit, yfit, w0
```

```python
## load example file - could also load many and average.  Ideally find quiet sun, definitely disk center

#filename='/Volumes/ViSP_External/pid_2_12/AEXOJ/VISP_2024_06_18T17_53_34_522_00589300_I_AEXOJ_L1.fits'
filename='/Volumes/ViSP_External/pid_2_12/BGVLL/VISP_2024_06_18T17_53_34_522_00630200_I_BGVLL_L1.fits' 
#filename = '/Users/coletamburri/Downloads/VISP_2024_06_18T17_53_34_522_00486100_I_ADPNE_L1.fits'
```

```python
# astropy fits handling

hdul = fits.open(filename)
data = hdul[1].data
plt.imshow(data[0])
```

```python
# header print-out to see useful variables
header = hdul[1].header
header
```

```python
# populate wavelength array... can also do using min and max (also in header) but result is the same

wlloc = hdul[1].header['CRVAL2']
wlpix = hdul[1].header['CRPIX2']
wldelt = hdul[1].header['CDELT2']
```

```python
print(wlloc)
print(wlpix)
print(wldelt)
```

```python
np.shape(data)
```

```python
wlarr = np.zeros(np.shape(data)[1])

wlarr[int(wlpix)] = wlloc

for j in range(int(wlpix)+1,np.shape(data)[1],1):
    wlarr[j] = wlarr[j-1] + wldelt
for j in range(int(wlpix),-1,-1):
    wlarr[j] = wlarr[j+1] - wldelt
```

```python
# example specrum... last index is the slice *along the slit* to take
plt.plot(wlarr,data[0,:,300])
```

```python
mean = np.mean(data[0],1)
```

```python
# can also plot the mean, but make sure averaging only over QS!
plt.plot(wlarr,mean)
```

```python
# just to visualize the max and min of the absorption lines used
# in the Fe I case I have (on April 9th 2025) just taken the two 630 lines because of a lack of reference lines in the window...
# but probably this is not good practice.  Get absorption lines from solar atlas ideally.
plt.plot(data[0,:,300])
plt.axvline(280)
plt.axvline(320)
plt.axvline(880)
plt.axvline(900)
```

```python
# define absorption lines using e.g. BASS2000 (but maybe that's not the most accurate? Pick your fav disk center, QS spectrum)

#absline1 = 485.97411 #fe I
#absline2 = 486.36434 #fe I

#absline1 = 589.995 #naId1
#absline2 = 589.591 #naId2

absline1 = 629.92245 #
absline2 = 630.5808 #???

##for fe I
# indices along wl array - essentially the "axvline" entries in the above
#low indices of telluric lines in spectrum - lower
indmins = [280,880]
#upper indces of telluric lines in spectrum - upper
indmaxs = [320,900]

#for fe I
#indmins = [360,850]
#indmaxs = [400,900]
```

```python
# this is the cell that finds the difference between measured and reported wavelength values

dispersion_range = np.array(wlarr)

# define locations of reference lines
shiftsel_obs1 = np.where((dispersion_range>dispersion_range[indmins[0]]) & 
                         (dispersion_range<dispersion_range[indmaxs[0]]))
shiftsel_obs2 = np.where((dispersion_range>dispersion_range[indmins[1]]) & 
                         (dispersion_range<dispersion_range[indmaxs[1]]))

wavesel_obs1 = np.take(dispersion_range,shiftsel_obs1)[0]
fluxsel_obs1 = np.take(mean,shiftsel_obs1)[0]

wavesel_obs2 = np.take(dispersion_range,shiftsel_obs2)[0]
fluxsel_obs2 = np.take(mean,shiftsel_obs2)[0]

pixsel_obs1 = np.arange(0, wavesel_obs1.shape[0])
pixsel_obs2 = np.arange(0, wavesel_obs2.shape[0])

xfit1, yfit1, w0_1 = fitline(pixsel_obs1,fluxsel_obs1)
xfit2, yfit2, w0_2 = fitline(pixsel_obs2,fluxsel_obs2)

# shift in lines
dw1 = w0_1-absline1
dw2 = w0_2-absline2

dt = absline2 - absline1

# correction in wavelengths of reference lines
corr1 = w0_1 + indmins[0]
corr2 = w0_2 + indmins[1]

obsdiff = corr2 - corr1

#change per nm
rat = dt/obsdiff
```

```python
# printout of the estimated dispersion (nm)
print(rat)
```
