```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
import os
```

```python
base = '/Volumes/VBI_External/archiveA1/'
savefolder = '/Volumes/VBI_External/rawmovie/'
```

```python
hdul = fits.open(base+'atst.ics.vbiRed.dc.vcc.xfer.173364.535311-1.fits.fz')
```

```python
for i in range(529600,537000,1):
#for i in range(534250,534300,5):
    file = 'atst.ics.vbiRed.dc.vcc.xfer.173364.'+str(i)+'-1.fits.fz'
    if os.path.isfile(base+file):
        hdul = fits.open(base+file)
        data = hdul[1].data[0]
        plt.imsave(savefolder+str(i)+'.jpg',data,cmap='grey',dpi=200)
        hdul.close()
```

```python
fig,ax=plt.subplots(dpi=200)
ax.pcolormesh(hdul[1].data[0])
ax.invert_yaxis()
```
