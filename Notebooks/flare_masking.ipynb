```python
import numpy as np
from scipy.spatial.distance import cdist

def kmeans(X, k, distance_metric='euclidean', max_iters=100, tol=1e-4):
    # Initialize centroids randomly from the data points
    centroids = X[np.random.choice(np.shape(X)[0], k,replace='False')]
    
    for i in range(max_iters):
        # Compute distances between data points and centroids
        distances = cdist(X, centroids, metric=distance_metric)
        
        # Assign clusters based on the closest centroid
        clusters = np.argmin(distances, axis=1)
        
        # Store old centroids to check for convergence
        old_centroids = centroids.copy()
        
        # Update centroids as the mean of the clusters
        for j in range(k):
            if np.any(clusters == j):  # Check if cluster is not empty
                centroids[j] = X[clusters == j].mean(axis=0)
        
        # Check for convergence
        if np.linalg.norm(centroids - old_centroids) < tol:
            break
            
    return centroids, clusters
```

```python
def kmeans_func(start,masknum,nsteps,startspace,endspace,obs_avg,flarearr,normalize,n_clusters,cutoff,line_low=caII_low,line_high=caII_high,metric='euclidean'):
   
    frame_line = obs_avg[(start+(masknum)*nsteps):(start+(masknum+1)*nsteps),startspace:endspace]
    maxframe = np.max(frame_line)
    minframe = np.min(frame_line)

    masklim = cutoff*minframe

    mask = np.copy(frame_line)
    mask[mask < masklim] = 0
    mask[mask > masklim] = 1

    maskinds = np.where(mask > .5)
    x_mask = maskinds[0]
    y_mask = maskinds[1]

    line_profiles = []
    
    for i in range(len(x_mask)):
        line_profiles.append(flare_arr[(start*(masknum+1))+x_mask[i],line_low:line_high,(start*(masknum+1))+y_mask[i]])
        
    normprofiles_line = []

    for i in range(len(x_mask)):
        line_norm = normalize(line_profiles[i])
    
        normprofiles_line.append(line_norm)

    X_train, X_test= train_test_split((normprofiles_line))

    km = sklearn.cluster.KMeans(n_clusters=n_clusters,init='random').fit(normprofiles_line)

    groups = km.predict(normprofiles_line)
                       
    labs = km.labels_
    
    return frame_line, mask, km, normprofiles_line, groups, labs, x_mask, y_mask
```
